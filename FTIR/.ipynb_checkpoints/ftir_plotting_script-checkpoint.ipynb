```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# List of sample files to import
# Format for each entry is
# ['filename (without .csv extension)', 'sample name -- to be displayed on graph']
samples = [
    ['silica_fume', 'Silica fume'],
    ['MgCO3_nH2O', 'Hydromagnesite'],
]

all_data = []

for s in samples:
    data = pd.read_csv(s[0] + '.csv', skiprows=1)
    data['T'] = data['%T'] / 100
    data['A'] = 1 - data['T']
    data['Sample'] = s[1]
    
    all_data.append(data)
    
df = pd.concat(all_data, axis=0, ignore_index=True)
```

```python
sns.set(font_scale = 2.5)
sns.set_style("white")
sns.set_palette(sns.color_palette("bright"))
plt.rcParams["font.family"] = "Times New Roman"
f, ax = plt.subplots(figsize=(15, 10))
ax.tick_params(bottom=True, left=True, width=2, length=10)

plot = sns.lineplot(
    data=df,
    hue='Sample',
    x="cm-1",
    y="T",
)

# Set x label and y label
plot.set_xlabel("Wavenumber (1/cm)")
plot.set_ylabel("Transmittance")

# Inverse x-axis so wavenumber is large -> small
ax.set_xlim(ax.get_xlim()[::-1])

# ------ Some optional graph settings that I frequently use -----

# Set x and y limits to zoom in on a certain section
# ax.set_xlim(1200, 900)
# ax.set_ylim(0.8, 1.0)

# Draw a vertical line at a specific wavenumber
# plt.axvline(x=3692, color='r', linestyle='--')

# Move lengend the the right side of the plot
# plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

# Save figure as svg to current directory
# I use Adobe Illustrator to add additional annotations
# plt.savefig('filename.svg')
```
